## Imports and Libraries

```python
import cv2 as cv
import os
import time
import winsound
from PIL import Image
```

# Global Variables

```python
# Change this to where you want to save your images!
savepath = "/Users/molte/OneDrive/Desktop/UAFS/~Spring 2024/Capstone/Unspoiled/images/"

classes = ["milk", "eggcarton", "creamer", "apple", "banana", "pear", "coughdrops", "cheezit", "soda"]
'''
classIDs:
0 = "milk"
1 = "eggcarton"
2 = "creamer"
3 = "apple"
4 = "banana"
5 = "pear"
6 = "coughdrops"
7 = "cheezit"
8 = "soda"
'''

##################################
# CHANGE THIS BEFORE TAKING PICS #
##################################
numPics = 100                   # number of pics to take
interval = 3                      # how many seconds to wait before taking another pic
classID = 4                       # index of item in classes[]

offset = 0                        # ONLY change this if you don't want to start at 0
base = classID * 10000 + offset   # what number to start the picture count at

cmap = "gray"
```

# Methods

```python
# Creates the necessary folder
def createFolder(item):
    
    if not os.path.exists(savepath + item):
        os.makedirs(savepath + item)
```

```python
# Takes a photo
def takePics(numPics, interval, item, base):

    for i in range(numPics):
        
        # Saves the image after a delay
        time.sleep(interval)
        
        picNum = base + i
        print(item, picNum, "...")

        # Takes a photo
        ret, img = camera.read()
    
        # Checks if the picture was taken
        if not ret:
            print("Error: Cannot capture frame from the camera.")
            continue

        # Writes the image to the savepath
        filename = item + "/" + item + str(picNum) + ".png"
        print(savepath + filename)
        cv.imwrite(savepath + filename, img)

        if i % 10 != 0:
            winsound.Beep(2500, 300)
        else:
            winsound.Beep(4000, 150)
            time.sleep(.01)
            winsound.Beep(4000, 150)
        
    print("done")
    
    winsound.Beep(1000, 150)
    time.sleep(.01)
    winsound.Beep(1000, 150)
    time.sleep(.01)
    winsound.Beep(1000, 150)
    time.sleep(.01)
    winsound.Beep(1000, 150)
    time.sleep(.01)
    winsound.Beep(1000, 150)
```

# Bot Code

```python
# Opens camera (sometimes takes longer than expected, at least for me)
camID = 1     # Built-in webcam is ID 0. You can change this to whatever your camera's ID is
camera = cv.VideoCapture(camID)

# Camera defaults to 640x480, this changes that to what it should be
camera.set(cv.CAP_PROP_FRAME_WIDTH, 1280)
camera.set(cv.CAP_PROP_FRAME_HEIGHT, 720)

print("CAMERA OPEN...") # This is just here so I know it finished this
```

# Taking Pictures

##### If camera is already open, just run this block again

```python
# Exits out if camera not open
if not camera.isOpened():
    print("CAMERA NOT OPEN...")
    print("OPEN IT FIRST...")
else:
    # Loops and takes pictures
    item = classes[classID]
    
    createFolder(item)
    takePics(numPics, interval, item, base)
```

```python
# Frees the camera from its technological shackles
camera.release()
```
